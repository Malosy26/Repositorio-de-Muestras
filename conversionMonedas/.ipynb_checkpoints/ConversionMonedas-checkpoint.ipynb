```python
elecciones = []
for i in range(2):
   
    if(i==0):
        print("Opciones disponibles:")
        for clave, valor in opciones.items():
            print(f"{clave}: {valor}")
        eleccion = input(f'Elija la opción {i+1} que desea comparar: ')
    else:    
        eleccion = input(f'Elija la opción {i+1} que desea comparar: ')
        
    if eleccion in opciones:
        print(f'Ha elegido: {eleccion} - {opciones[eleccion]}')
        elecciones.append(eleccion)
    else:
        print('La opción elegida no es válida')
        break
cantidad=int(input("Introduzca en un entero la cantidad deseada a convertir"))
```

```python
print(f'Usted va a recibir la grafica de la conversion de la cantidad de {cantidad} {elecciones[0]} a {elecciones[1]} desde el principio de 2023 hasta hoy' )
```

```python
#CONFIGURACION LA PETICION DE DATOS

from_currency=elecciones[0]
to_currency = elecciones[1]
date_1 = "2023-01-01"
endpoint = f"{url}/{date_1}..?amount={cantidad}&from={from_currency}&to={to_currency}"
response = requests.get(endpoint)
data=response.json()
hazGrafica(data,elecciones)
#He tenido a bien hacer una funcion para la grafica a fin de separar cada aplicacion
    
```

```python
def hazGrafica(data,monedas)->None:
    #CONVERTIR LAS FECHAS A OBJETO DATETIME 
    fechas = list(data["rates"].keys())
    fechas = [datetime.strptime(fecha, "%Y-%m-%d") for fecha in fechas]

    
    monedasCon = [data["rates"][fecha][monedas[1]] for fecha in data["rates"].keys()]
  
    
    plt.figure(figsize = (18, 10))
    plt.plot_date(x = fechas, y = monedasCon, fmt = "--")

    # Evita que se sobrepongan las fechas
    plt.gcf().autofmt_xdate()

  # max    np.argmax(monedasCon) devuelve el indice donde esta el valor maximo
    plt.plot(fechas[np.argmax(monedasCon)],
             np.max(monedasCon),
             color      = "black",
             marker     = "*",
             markersize = 10,
             label      = f"max: {np.max(monedasCon)}")

    # min
    plt.plot(fechas[np.argmin(monedasCon)],
             np.min(monedasCon),
             color      = "black",
             marker     = "*",
             markersize = 10,
             label      = f"min: {np.min(monedasCon)}")

    
    
    
    
    # Etiquetas
    plt.legend()
    plt.xlabel("Fechas", fontsize = 20)
    plt.ylabel(f'{elecciones[0]}', fontsize = 20)
    plt.title(f'Relación de 1 {monedas[0]} - 1 {monedas[1]} desde 01/01/2023 / {np.max(fechas).strftime("%d-%m-%Y")}', fontsize = 20)


    plt.grid()
    plt.show()
    
```
